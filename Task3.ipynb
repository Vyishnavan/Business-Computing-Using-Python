```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1
#load the dataset using url
college_major = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
```

```python
college_major.head(10)
```

```python
# 2
# Calculate the correlation between Median salary and Unemployment_rate
college_major['Median'].corr(college_major['Unemployment_rate'])
```

```python
# Scatter plot between Median salary and Unemployment_rate
college_major.plot(kind='scatter',title='Scatter plot for Median salary and Unemployent_rate ', x='Median', y='Unemployment_rate')
plt.show()
```

```python
# 3
#Find majors where the Median salary is above 60000 and the Unemployment_rate is below 5%
college_major[(college_major["Median"] > 60000) & (college_major["Unemployment_rate"] < 0.05)]
```

```python
# 4
# Group the data by Major_category and calculate the average Median salary for each category
grouped_data = college_major.groupby("Major_category")
grouped_data_mean = grouped_data["Median"].mean()
print(grouped_data_mean)
```

```python
grouped_data_mean.plot(kind='bar')
plt.xlabel('Major_category')
plt.ylabel('Average Median salary')
plt.title('Average Median salary for each category')
plt.show()
```
