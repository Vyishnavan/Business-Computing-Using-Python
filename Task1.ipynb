```python
# importing pandas
import pandas as pd
```

```python
# load the Data
#Q1
df = pd.read_csv("car_price_dataset.csv")
```

```python
#Q2
# Using the describe method to get the mean, minimum, maximum and standard deviation 
df.describe()
```

```python
#Q3
#Calculate the median of Mileage
df['Mileage'].median()
```

```python
#Calculate the median of Price
df['Price'].median()
```

```python
# importing library to create a scatter plot
import matplotlib as mpl
```

```python
#Q4
#Create a scatter plot with Mileage(y-axis) and Price(x-axis)
df.plot(kind='scatter',title='Scatter plot for Price and Mileage ', x='Price', y='Mileage')
mpl.pyplot.show()
```

```python
#Q 4.2 The scatter plot shows a negative relationship between Mileage and Price. Higher Mileage cars tend to have lower prices, which suggests  that as the usage of the car increases, its value decreases.
```

```python
#Q5
#Calculate the correlation betwwen Mileage and Price
df['Mileage'].corr(df['Price'])
```

```python
#Calculate the correlation betwwen Engine_Size and Price
df['Engine_Size'].corr(df['Price'])
```

```python
#Q 5.2 The correlation between mileage and price is negative, indicating that higher mileage is associated with lower prices. On the other side, the correlation between engine size and price is positive, suggesting that cars with larger engine sizes tend to have higher prices.
```
