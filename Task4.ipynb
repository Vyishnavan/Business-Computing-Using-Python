```python
import sqlite3 as sq3
```

```python
# Q1
# Create the connection
conn = sq3.connect('Hospital_System.db')
```

```python
#SQL query to create a table for Hospital
sql = ''' CREATE TABLE Hospital(
hospital_id INT(3),
address VARCHAR(50) UNIQUE NOT NULL,
zip_code VARCHAR(10) NOT NULL,
city VARCHAR(50) NOT NULL,
region VARCHAR(50) NOT NULL,
country VARCHAR(50) NOT NULL,
CONSTRAINT hospital_pk PRIMARY KEY (hospital_id));'''
```

```python
conn.execute(sql)
```

```python
#SQL query to create a table for Wards
sql = ''' CREATE TABLE Wards(
ward_id INT(4),
ward_name VARCHAR(20) UNIQUE NOT NULL,
hospital_id INT(3),
CONSTRAINT ward_pk PRIMARY KEY(ward_id),
CONSTRAINT ward_fk FOREIGN KEY(hospital_id)
REFERENCES branches(ward_id));'''
```

```python
conn.execute(sql)
```
