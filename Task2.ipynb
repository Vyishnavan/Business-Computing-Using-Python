```python
# For loop
length_category = []
for word_count in df['word_count']:
    if word_count < 100:
        length_category.append("short")
    elif word_count > 200:
        length_category.append("long")
    else:
        length_category.append("medium")
        
        
```

```python
df['length_category'] = length_category
```

```python
df
```

```python
#while loop
length_category = []
```

```python
index = 0
while index < len(df):
    word_count = df.loc[index, 'word_count']
    if word_count < 100:
        length_category.append("short")
    elif word_count > 200:
        length_category.append("long")
    else:
        length_category.append("medium")
    index += 1
```

```python
df['length_category'] = length_category
```

```python
#Vectorisation
df['length_category'] = df['word_count'].apply(lambda word_count: 'short' if word_count <100 else('long' if word_count > 200 else 'medium'))
```
